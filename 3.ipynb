```python

from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (x**2 - 1 - y, np.exp(x) + x * (1 - y) - 1)

x1, y1 =  fsolve(equations, (3, 8))
x2, y2 =  fsolve(equations, (-2, -3))
              
print (x1, y1)
print (x2, y2)
```
